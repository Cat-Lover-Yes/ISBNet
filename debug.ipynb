```python
import torch
import os
import numpy as np
import glob

# file = '/home/ubuntu/3dis_ws/ISBNet/dataset/scannet200/train/scene0000_00_inst_nostuff.pth'

# xyz, rgb, sem_id, inst_ind = torch.load(file)
# breakpoint()

files = [torch.load(f) for f in glob.glob('/home/ubuntu/3dis_ws/ISBNet/dataset/scannet200/train/*.pth')]


```

```python
totals = 0
count = np.zeros((200), dtype=np.long)

for file in files:
    xyz, rgb, sem_id, inst_ind = file

    sem_id_ = sem_id[sem_id!=-100]
    totals += len(sem_id_)
    count_ = np.bincount(sem_id_, minlength=200)

    count += count_
```

```python
print(totals)
print(count)
```

```python
weight = totals / count

weight_norm = weight/weight[1]
```

```python
print(weight_norm)
```
